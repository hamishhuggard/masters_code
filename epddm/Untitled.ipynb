```python
import seaborn as sns
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time", row="sex",
            hue="smoker", style="smoker", size="size",
            data=tips, kind='line');
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(tips.total_bill, tips.tip, '.r');
```
